<h2>Importing the libraries </h2>

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.layers import Convolution2D,MaxPooling2D,Flatten,Dense
from tensorflow.keras.models import Sequential
import cv2 as cv
```

```python
!unzip "/content/drive/MyDrive/Bird_species.zip"
```

```python
train_gen = ImageDataGenerator(rescale = (1./255),horizontal_flip=True,shear_range=0.2,zoom_range=0.2)
test_gen = ImageDataGenerator(rescale = (1./255))
```

```python
train=train_gen.flow_from_directory('/content/train_data/train_data',
                                    target_size=(120,120),
                                    class_mode='categorical',
                                    batch_size=8)
test=test_gen.flow_from_directory('/content/test_data/test_data',
                                  target_size=(120,120),
                                    class_mode='categorical',
                                    batch_size=8)
```

```python
model = Sequential([
    Convolution2D(20,(3,3),activation = 'relu',input_shape=(120,120,3)),
    MaxPooling2D(2,2),
    Flatten(),
    Dense(45,activation = 'relu'),
    Dense(16,activation = 'softmax')
])

model.summary()
```

```python
train.class_indices
```

```python
model.compile(optimizer='adam',metrics=['accuracy'],loss='categorical_crossentropy')
```

```python
model_fit = model.fit(train,epochs =20,
                     validation_data = test ,batch_size=5)
```

```python
model.save('birds.h5')
```

# Model Tuning(with Drop out , batch normalisation , early stopping)


```python
from tensorflow.keras.layers import BatchNormalization,Dropout
```

```python
model = Sequential([
    Convolution2D(12,(3,3),activation = 'relu',input_shape=(120,120,3)),
    BatchNormalization(),
    MaxPooling2D(2,2),
    Dropout(0.2),
    Convolution2D(24,(3,3),activation = 'relu'),
    BatchNormalization(),
    MaxPooling2D(2,2),
    Dropout(0.2),
    Convolution2D(36,(3,3),activation = 'relu'),
    BatchNormalization(),
    MaxPooling2D(2,2),
    Dropout(0.2),
    Flatten(),
    Dense(62,activation = 'relu'),
    BatchNormalization(),
    Dropout(0.1),
    Dense(32,activation = 'relu'),
    Dense(16,activation = 'relu'),
    Dense(16,activation = 'softmax')
])
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',metrics=['accuracy'],loss='categorical_crossentropy')
```

```python
# Early stopping
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor = 'val_accuracy',patience = 8)
```

```python
model_fit = model.fit(train,epochs =30,
                     validation_data = test ,
                      batch_size=5,
                      callbacks=early_stop)
```
